```python
# Clean up game scores
import datetime

# Convert into array of scores
def convert_to_score_array(val):
    # Split by space and append to array
    arr = []
    if type(val) == str:
        for score in val.split(", "):
            arr.append(score)
    elif type(val) == datetime.datetime:
        arr.append(val)
    return arr

# Convert timestamp to game score
def convert_to_game_score(values):
    temp = []
    for val in values:
        if type(val) == datetime.datetime:
            val = pd.Timestamp(val)
            # Extract month and date
            month = val.month
            day = val.day
            temp.append((int(month), int(day)))
        elif type(val) == str:
            a, b = val.split("-")
            temp.append((int(a), int(b)))
            
    return temp
```

```python
df_subset["Game Scores"] = df_subset["Game Scores"].apply(convert_to_score_array).apply(convert_to_game_score)
```

```python
# List of unique players
unique_a1 = set(list(df_subset["Player A1 Squash BC Number"].unique()))
unique_a2 = set(list(df_subset["Player A2 Squash BC Number"].unique()))
unique_b1 = set(list(df_subset["Player B1 Squash BC Number"].unique()))
unique_b2 = set(list(df_subset["Player B2 Squash BC Number"].unique()))

overall_unique_players = unique_a1.union(unique_a2).union(unique_b1).union(unique_b2)

print(str(len(list(df_subset["Player A1 Squash BC Number"].unique()))) + " unique Player A1")
print(str(len(list(df_subset["Player A2 Squash BC Number"].unique()))) + " unique Player A2")
print(str(len(list(df_subset["Player B1 Squash BC Number"].unique()))) + " unique Player B1")
print(str(len(list(df_subset["Player B2 Squash BC Number"].unique()))) + " unique Player B2")
print(str(len(overall_unique_players)) + " unique players")
```

```python
# List of unique genders
unique_a1_gender = set(list(df_subset["Player A1 Gender"].unique()))
unique_a2_gender = set(list(df_subset["Player A2 Gender"].unique()))
unique_b1_gender = set(list(df_subset["Player B1 Gender"].unique()))
unique_b2_gender = set(list(df_subset["Player B2 Gender"].unique()))

overall_unique_player_genders = unique_a1_gender.union(unique_a2_gender).union(unique_b1_gender).union(unique_b2_gender)

print(str(len(list(df_subset["Player A1 Gender"].unique()))) + " unique Player A1 genders")
print(str(len(list(df_subset["Player A2 Gender"].unique()))) + " unique Player A2 genders")
print(str(len(list(df_subset["Player B1 Gender"].unique()))) + " unique Player B1 genders")
print(str(len(list(df_subset["Player B2 Gender"].unique()))) + " unique Player B2 genders")
print(str(len(overall_unique_player_genders)) + " unique genders")
print(overall_unique_player_genders) # nan, female, male -> unspecifed, female, male
```

```python
# Convert "nan" genders to "unspecified"

df_subset["Player A1 Gender"].fillna("unspecified", inplace=True)
df_subset["Player A2 Gender"].fillna("unspecified", inplace=True)
df_subset["Player B1 Gender"].fillna("unspecified", inplace=True)
df_subset["Player B2 Gender"].fillna("unspecified", inplace=True)
```
